#### 2. List out all available cuisine from the whole dataset and sort them in alphabetical order.

#### 3. Find out all available cuisine from the restaurants that are located on Cross Bay Boulevard and whose address uses zip code 11414.

```python
[d for d in col.find({'address.zipcode': "11414", 'address.street': 'Cross Bay Boulevard'})]
```

#### 4. Find the name and address of the restaurants containing Steak House in the name (Hint: use regular expression).


```python
import re
```

```python
regx= re.compile('steak house', re.IGNORECASE)
[d for d in col.find({'name':{'$regex':regx}})]
```

#### 5. List out the name of all restaurants which contain the word Pizza in the cuisine but DO NOT contain the word Pizza or Pizzeria in the restaurant name (Hint: use regular expression).

```python
regex= re.compile('pizza', re.IGNORECASE)
regex2= re.compile('pizza|pizzeria', re.IGNORECASE)
[d for d in col.find({'cuisine': {'$regex': regex}, 'name': {'$not': {'$regex': regex2}}})]
```

#### 6. List out the name of all straight A (i.e. the restaurant has only received A grade ever) restaurants which contain the word Pizza in the cuisine and are located in the Queens borough (Hint: first find out how many available grade values we have in the entire dataset).

```python
[d for d in col.find({'grades': {'$elemMatch': {'grade': 'A'}},'cuisine': {'$regex': 'pizza', '$options': 'i'}, 'borough' : 'Queens'})]
```

#### 7. Find out  the number of restaurants listed Hamburgers as their main cuisine.

```python
col.count_documents({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}})
```

#### 8. You are in Manhattan right now so let's find how many restaurants listed Hamburgers as their main cuisine in the Manhattan borough.

```python
col.count_documents({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'borough': 'Manhattan'})
```

#### 9. Find out how many restaurants listed Hamburgers as their main cuisine in Manhattan and exclude all Mcdonald's (Note: In the data set, McDonald's was presented in inconsistent ways, e.g. McDonald's and McDonald'S. Use the regular expression /McDonald/ in your query).

```python
col.count_documents({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'name': {'$not': {'$regex': 'mcdonald', '$options': 'i'}}, 'borough': 'Manhattan'})#col.count_documents({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'name': {'$regex': 'mcdonald', '$options': 'i'}, 'borough': 'Manhattan'})
#[d for d in col.find({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'name': {'$not': {'$regex': 'mcdonald', '$options': 'i'}}, 'borough': 'Manhattan'})]
```

#### 10. Exclude also Burger King in the previous query.

```python
regex= re.compile('mcdonald|burger king', re.IGNORECASE)
col.count_documents({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'name': {'$not': {'$regex': regex}}, 'borough': 'Manhattan'})
#[d for d in col.find({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'name': {'$not': {'$regex': regex}}, 'borough': 'Manhattan'})]
```

#### 11. Find out the list of distinct street based on the previous results.

```python
#[d for d in col.find({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'name': {'$not': {'$regex': regex}}, 'borough': 'Manhattan'})]
col.find({'cuisine': {'$regex': 'hamburgers', '$options': 'i'}, 'name': {'$not': {'$regex': regex}}, 'borough': 'Manhattan'}).distinct('address.street')
```
